#**Matplotlib Basics**

Matplotlib is a great 2D plotting library for python. This notebook contains some basic samples for getting started with matplotlib.

```
# import the Matplotlib module
import matplotlib.pyplot as plt

# import the numpy module
import numpy as np
```

The matplotlib chart is created  using the pyplot in the matplotlib module . To make it simple we have imported it as plt.

To create a chart define a list of values for x and y axis and pass it inside the plot function.

show() functions diplays the chart.

```
# simple linear chart

# chart value for both the axes
a = [1, 2, 3]

# plot the graph
plt.plot(a)
plt.show()
```

### Chart title
The title can be added using the title() function, which takes the string as a parameter.

```
# display the chart title

plt.title("Simple linear chart")
plt.plot(a)
plt.show()
```

## X and Y axis label
xlabel and ylabel functions are used

```
# display labels for x and y axis

plt.title("Simple linear chart")

plt.xlabel("X axis")   # x axis label
plt.ylabel("Y axis")   # y axis label

plt.plot(a)
plt.show()
```

## Changing the fontsize of the labels and title

```
# increase title and labels font size

plt.title("Simple linear chart", fontsize=20)
plt.xlabel("X axis", fontsize=15)
plt.ylabel("Y axis", fontsize=15)
plt.plot(a)
plt.show()
```

## Displaying different X and Y values

```
# plotting different data on both axis

x_data = [1, 2, 3, 5]
y_data = [4, 6, 5, 2]

plt.plot(x_data, y_data)
plt.show()
```

```
# change the axis values

x_data = [1, 2, 3, 5, 20]
y_data = [4, 6, 5, 2, 1]

# axis values [xmin, xmax, ymin, ymax]
# try changes these values
plt.axis = [0, 10, 0, 20]

plt.plot(x_data, y_data)
plt.show()
```

## Line properties
[Click here for more info about line properties](https://matplotlib.org/tutorials/introductory/pyplot.html#controlling-line-properties)

```
# change the line color

x_data = [1, 2, 3, 5]
y_data = [4, 6, 5, 2]

plt.plot(x_data, y_data, 'r') # 'r' means red, try 'g' and 'b' for green and blue
plt.show()
```

```
# formatting the line width

x_data = [1, 2, 3, 5]
y_data = [4, 6, 5, 2]

plt.plot(x_data, y_data, linewidth=5.0)
plt.show()
```

```
# formatting the line type

x_data = [1, 2, 3, 5]
y_data = [4, 6, 5, 2]

plt.plot(x_data, y_data, 'r-') # '-' solid line 'o' dot (concat the color string with the line style string)
plt.plot(x_data, y_data, '>') 
plt.show()
```

```
# plot more than one line in the same chart

x1, y1 = [1, 2, 3, 4], [1, 2, 3, 4] # line 1
x2, y2 = [4, 3, 2, 1], [1, 2, 3, 4] # line 2

# plt.plot(x1, y1, x2, y2) # without line style
plt.plot(x1, y1, 'r--', x2, y2, 'g-') # with line style
plt.show()
```

```
# controlling line properties using the setp function

x1, y1 = [1, 2, 3, 4], [1, 2, 3, 4] # line 1
x2, y2 = [4, 3, 2, 1], [1, 2, 3, 4] # line 2
lines = plt.plot(x1, y1, x2, y2)

# use keyword args
plt.setp(lines, color='b', linewidth=3.0, linestyle='--')
plt.show()
```

## Plotting with categoriacal values


```
# create categorical data

cat_names = ['cat_1', 'cat_2', 'cat_3', 'cat_4']
cat_values = [20, 100, 45, 55]
```

```
# bar chart

plt.bar(cat_names, cat_values)
plt.show()
```

```
# scatter

plt.scatter(cat_names, cat_values)
plt.show()
```

```
# histogram

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')

# add text to the chart
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')

# show grid
plt.grid(True) # try changing this values to False

plt.show()
```

## Multiple figures and subplots

```
# create figures
plt.figure(1) # first figure

# create subplot ex. subplot(nrows, ncols, index, **kwargs)
plt.subplot(1, 2, 1)        # subplot with one row and 2 columns
plt.plot([1, 2, 3, 4])      # plot in the first index

plt.subplot(1, 2, 2)        # plot in the second index
plt.plot([4, 3, 2, 1])

# create second figure
plt.figure(2)

plt.subplot(111)             # 1, 1, 1 can be written as 111 in subplot
plt.plot([4, 3, 2, 1])       # plot in the second figure

plt.title("Second figure")

plt.show()
```

## Saving chart to file

```
# save the chart as a file

x_data = [1, 2, 3, 5]
y_data = [4, 6, 5, 2]

plt.plot(x_data, y_data, 'r')

# save in pdf formats
plt.savefig('sample.pdf', format='pdf')
```
